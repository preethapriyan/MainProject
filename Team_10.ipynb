Loop for outlier removal

```python
numeric_col= ['Temperature(F)','Wind_Chill(F)','Wind_Speed(mph)']
```

```python
#To find outlier in Temperature , Wind_Chill(F),Wind_Speed(mph) column
for y in numeric_col:
    #Calculating Q1, Q2, Q3 values
    Q1=np.percentile(data[y],25,interpolation='midpoint')
    Q2=np.percentile(data[y],50,interpolation='midpoint')
    Q3=np.percentile(data[y],75,interpolation='midpoint')
    print(Q1)
    print(Q2)
    print(Q3)
    if(y =="Temperature(F)"):
       limit=3
    if(y =="Wind_Speed(mph)"):
       limit=7
    if(y =="Wind_Chill(F)"):
       limit=7
    print(limit)
    #Calculating IQR, lower, upper values
    IQR=Q3-Q1
    low_lim=(Q1-(limit*IQR))
    Up_lim=(Q3+(limit*IQR))
    print("Outlier Lower Limit : ",low_lim)
    print("Outlier Upper Limit: ",Up_lim)
    #Determining the outliers
    outlier=[]
    for x in data[y]:
        if((x>Up_lim) or (x<low_lim)):
            outlier.append(x)
    print(len(outlier))
    print("Outlier : ",outlier)
    #Finding Index of outliers
    index1=data[y]>Up_lim
    ind=[]
    ind=data.loc[index1].index
    print(ind)
    #Removing outliers
    for i in ind:
        data.drop(i,inplace=True)
       
data.boxplot(numeric_col)
data.shape
```

So we had deleted 9 + 3778 + 894 = 4681 rows as part of outlier detection.

```python
numeric_col4= [ 'Pressure(in)','Distance(mi)','Visibility(mi)']
```

```python
#To find outlier in  Pressure(in), Distance(mi), Visibility(mi)   column
for y in numeric_col4:
    #Calculating Q1, Q2, Q3 values
    Q1=np.percentile(data[y],25,interpolation='midpoint')
    Q2=np.percentile(data[y],50,interpolation='midpoint')
    Q3=np.percentile(data[y],75,interpolation='midpoint')
    print(Q1)
    print(Q2)
    print(Q3)
  
    #Calculating IQR values
    IQR=Q3-Q1
    print("IQR= ",IQR)
```

It is very clear that for the columns Pressure(in), Distance(mi), Visibility(mi), the IQR value is almost zero, therefore discarding the outliers of these columns.

```python
#Linear Regression
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,r2_score,mean_squared_error
mse=mean_squared_error(y_test,y_pred)
r2_score=r2_score(y_test,y_pred)
print("Linear Regression:\n"'R2 score:',r2_score)
print('Mean Squared error:',mse)
```

```python
#To display accuracy
def dispAcc(y_pred):
  print("Classification Report")
  print(classification_report(y_test,y_pred))
  print("Accuracy is ",accuracy_score(y_test,y_pred))
  print("Precision is ", precision_score(y_test,y_pred,pos_label='positive',
                                           average='micro'))
  print("Recall is ",recall_score(y_test,y_pred,y_pred,pos_label='positive',
                                           average='micro'))
  print("f1 score is ",f1_score(y_test,y_pred,y_pred,pos_label='positive',
                                           average='micro'))
  confusion_matrix(y_test,y_pred)
```
